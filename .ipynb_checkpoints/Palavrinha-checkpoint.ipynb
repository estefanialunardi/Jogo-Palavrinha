```python
def palavra_do_usuario (random_list, tentativa):
    """ Verifica se a palavra escolhida pelo usuário corresponde à randômica,
    dando dicas se a letra existe na palavra (em amarelo) e se está na posição correta
    (em verde). Caso a letra não exista, será impressa em vermelho.
    Arg. random_list, tentativa
    Return: resultado"""
    
    chosen_word = input("Sou uma palavra de cinco letras")
    chosen_word=list(chosen_word)
    if tentativa<5:
        if len(chosen_word)!=5:
            print("Eu tenho cinco letras! Tente de novo!")
            chosen_word = palavra_do_usuario (random_list, tentativa)
        else:
            tentativa+=1
            palavrinha=[]
            cor=[]
            index=0
            chosen_word=list(chosen_word)
            for letra in chosen_word:
                index+=1
                if letra in random_list:
                    if random_list[index-1] == chosen_word[index-1]:
                            palavrinha.append(letra)
                            cor.append(2)
                    else:
                        if random_list.count(letra)>=chosen_word.count(letra):
                            if letra not in palavrinha:
                                palavrinha.append(letra)
                                cor.append(1)
                            else:
                                palavrinha.append(letra)
                                cor.append(1)
                        else:
                            palavrinha.append(letra)
                            cor.append(0)
                else:
                    palavrinha.append(letra)
                    cor.append(0)
               
        impressao=[]
        class bcolors:
            GREEN = '\033[32m' 
            YELLOW = '\033[93m'
            RED = '\033[91m' 
            RESET = '\033[0m'
        rodada=0
        for letter in palavrinha:
            rodada+=1
            
            if cor[rodada-1]==1:
                impressao.append(f"{bcolors.YELLOW}{letter}{bcolors.RESET}")
            elif cor[rodada-1]==0:
                impressao.append(f"{bcolors.RED}{letter}{bcolors.RESET}")
            elif cor[rodada-1]==2:
                impressao.append(f"{bcolors.GREEN}{letter}{bcolors.RESET}")
        impressao="".join(impressao)
        print(impressao)     
        
        if cor == [2, 2, 2, 2, 2]:
            print (f"Parabéns! Você acertou a palavra em {tentativa} tentativas!")
        else:
            print (f"Você ainda não acertou. Ainda tem {6-tentativa} tentativas!")
            chosen_word = palavra_do_usuario (random_list, tentativa)
    else:
        print (f"Você não conseguiu desta vez! A palavra era {random_word} Tente novamente")
    #return chosen_word
```
