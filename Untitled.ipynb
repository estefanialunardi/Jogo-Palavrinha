```python
tentativa=0
def palavra_do_usuario (random_list, tentativa):
    chosen_word = input("Sou uma palavra de cinco letras")
    while tentativa<5:
        chosen_word=list(chosen_word)
       #dict_chosen={}
       #dict_random={}
       #for caract in chosen_word:
       #    dict_chosen[caract]=(chosen_word.count(caract))
       #print(dict_chosen)
       #for caract in random_list:
       #    dict_random[caract]=(random_list.count(caract))
       #print(dict_random)
        if len(chosen_word)!=5:
            print("Eu tenho cinco letras! Tente de novo!")
            palavra_do_usuario (random_list, tentativa)
        #elif "".join(chosen_word) not in lista_randomica:
           #print(chosen_word)
            #print("Não conheço esta palavra... Tente de novo!")
            #palavra_do_usuario (random_list)
        else:
            tentativa+=1
            palavrinha=[]
            cor=[]
            index=0
            chosen_word=list(chosen_word)
            for letra in chosen_word:
                index+=1
                if letra in random_list:
                    if random_list.count(letra)>=chosen_word.count(letra):
                        if random_list[index-1] == chosen_word[index-1]:
                            palavrinha.append(letra)
                            cor.append(2)                       
                        else:
                            palavrinha.append(letra)
                            cor.append(1)
                    else:
                        if letra not in palavrinha:
                            if random_list[index-1] == chosen_word[index-1]:
                                palavrinha.append(letra)
                                cor.append(2)                       
                            else:
                                palavrinha.append(letra)
                                cor.append(1)
                        else:
                            palavrinha.append(letra)
                            cor.append(0)
                else:
                    palavrinha.append(letra)
                    cor.append(0)
               
        impressao=[]
        class bcolors:
            GREEN = '\033[32m' #GREEN-============
            YELLOW = '\033[93m' #YELLOW
            RED = '\033[91m' #RED
            RESET = '\033[0m'
        rodada=0
        for letter in palavrinha:
            rodada+=1
            
            if cor[rodada-1]==1:
                impressao.append(f"{bcolors.YELLOW}{letter}{bcolors.RESET}")
            elif cor[rodada-1]==0:
                impressao.append(f"{bcolors.RED}{letter}{bcolors.RESET}")
            elif cor[rodada-1]==2:
                impressao.append(f"{bcolors.GREEN}{letter}{bcolors.RESET}")
        impressao="".join(impressao)
        resultado = print(impressao)     
        
        if cor == [2, 2, 2, 2, 2]:
            resultado = print (f"Parabéns! Você acertou a palavra em {tentativa} tentativas!")
            break
        else:
            resultado = print (f"Você ainda não acertou. Ainda tem {6-tentativa} tentativas!")
            palavra_do_usuario (random_list, tentativa)
    else:
        resultado = print (f"Você não conseguiu desta vez! A palavra era {random_word} Tente novamente")
    return resultado
```
